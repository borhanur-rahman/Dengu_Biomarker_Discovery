```python
import scanpy as sc
import pandas as pd
import glob
import os
from tqdm import tqdm
import scipy.sparse as sparse

folder_path = "/content/"
tsv_files = glob.glob(os.path.join(folder_path, "*_counts.tsv.gz"))

print(f"Total files: {len(tsv_files)}")

batch_size = 1000  # ১০০০ করে লোড (তুমি চাইলে 500 করতে পারো যদি error আসে)
adata_full = None

for i in tqdm(range(0, len(tsv_files), batch_size)):
    batch_files = tsv_files[i:i + batch_size]
    adata_batch_list = []

    for file_path in batch_files:
        gsm_id = os.path.basename(file_path).split('_')[0]
        try:
            df = pd.read_csv(file_path, sep='\t', index_col=0, compression='gzip')
            if df.empty:
                continue
            df = df.T  # cells × genes

            # unique cell name
            cell_names = [f"{gsm_id}_{idx}" for idx in df.index]
            df.index = cell_names

            # sparse matrix-এ কনভার্ট (memory save)
            X_sparse = sparse.csr_matrix(df.values)

            ad = sc.AnnData(
                X=X_sparse,
                obs=pd.DataFrame(index=df.index),
                var=pd.DataFrame(index=df.columns)
            )
            ad.obs['sample_id'] = gsm_id
            ad.obs['cell_id'] = df.index
            adata_batch_list.append(ad)
        except Exception as e:
            print(f"Error in batch {gsm_id}: {e}")

    if adata_batch_list:
        adata_batch = sc.concat(adata_batch_list, join='outer')
        if adata_full is None:
            adata_full = adata_batch
        else:
            adata_full = sc.concat([adata_full, adata_batch], join='outer')

        print(f"Batch {i//batch_size + 1} done. Current cells: {adata_full.shape[0]}")

# Final cleanup
adata_full.var_names_make_unique()
adata_full.obs_names_make_unique()

print("\nFull Dataset Loaded!")
print("Total cells × genes:", adata_full.shape)
print("Unique samples:", adata_full.obs['sample_id'].nunique())
print("Cell counts per sample:\n", adata_full.obs['sample_id'].value_counts().head(10))
```

```python
# Approximate mapping (paper থেকে)
severe_prefixes = ['GSM3245112', 'GSM3245113', 'GSM3245114', 'GSM3245115']  # severe
non_severe_prefixes = ['GSM3245116', 'GSM3245117']  # dengue fever
healthy_prefixes = []  # healthy GSM গুলো যদি থাকে, যেমন GSM3245xxx healthy

adata_full.obs['condition'] = 'unknown'

# Prefix দিয়ে label দাও
adata_full.obs['condition'] = adata_full.obs['sample_id'].apply(
    lambda x: 'severe' if any(p in x for p in severe_prefixes) else
              'non-severe' if any(p in x for p in non_severe_prefixes) else
              'healthy' if any(p in x for p in healthy_prefixes) else 'unknown'
)

print("Condition distribution:")
print(adata_full.obs['condition'].value_counts())

# যদি সব 'unknown' হয় তাহলে বলো — আমরা আরও mapping খুঁজবো
```

```python
import scanpy as sc

print("Light Preprocessing শুরু হচ্ছে...")
print("Original shape:", adata_full.shape)

# ১. Basic filtering (আগের মতো)
sc.pp.filter_cells(adata_full, min_genes=200)
sc.pp.filter_genes(adata_full, min_cells=3)

print("After basic filter:", adata_full.shape)
```

```python
# ২. Mitochondrial QC (শুধু ক্যালকুলেট, ফিল্টার না)
adata_full.var['mt'] = adata_full.var_names.str.startswith('MT-') | adata_full.var_names.str.startswith('mt-')
sc.pp.calculate_qc_metrics(adata_full, qc_vars=['mt'], inplace=True)

print("\nQC Quick Stats:")
print("Mean genes per cell:", adata_full.obs['n_genes_by_counts'].mean())
print("Mean total counts:", adata_full.obs['total_counts'].mean())
print("Mean mt%:", adata_full.obs['pct_counts_mt'].mean())
```

```python
# ৩. Normalization + log (এটাই মূল)
sc.pp.normalize_total(adata_full, target_sum=1e4)
sc.pp.log1p(adata_full)

print("Normalization & log complete")
```

```python
# ৪. Highly variable genes (plot ছাড়া)
sc.pp.highly_variable_genes(adata_full, n_top_genes=2000, min_mean=0.0125, max_mean=3, min_disp=0.5)

print("Number of HVGs:", adata_full.var['highly_variable'].sum())

# Scaling skip করলাম (RAM save করার জন্য)
# যদি পরে লাগে তাহলে শুধু HVGs দিয়ে PCA/UMAP করা যাবে

print("\nLight Preprocessing সম্পূর্ণ!")
print("Final shape:", adata_full.shape)
print("Ready for Clustering (স্টেপ ৩)")

# যদি এখনো crash হয় তাহলে বলো — আমরা আরও ছোট করে করবো (যেমন 5000 cell নিয়ে)
```

```python
# Top 20 genes-এর symbol বের করার সঠিক উপায়
top_ensg = top_cluster5['names'].head(20).tolist()

# যদি var_names-এ gene symbol থাকে (যেমন 'MX2', 'IFIT1') তাহলে সরাসরি
if adata_full.var_names[0].startswith('ENSG'):  # যদি ENSG ID হয়
    print("var_names হচ্ছে ENSG ID — gene symbol পাওয়া যাবে না।")
    print("Top 20 ENSG IDs:")
    for i, ensg in enumerate(top_ensg, 1):
        print(f"{i}. {ensg} (logFC: {top_cluster5.loc[i-1, 'logfoldchanges']:.3f})")
else:
    # যদি var_names হয় gene symbol
    print("Top 20 genes (symbol + logFC):")
    for i, gene in enumerate(top_ensg, 1):
        logfc = top_cluster5.loc[i-1, 'logfoldchanges']
        print(f"{i}. {gene} (logFC: {logfc:.3f})")

# যদি gene_symbols column থাকে (কখনো কখনো থাকে)
if 'gene_symbols' in adata_full.var.columns:
    print("\nGene symbols column পাওয়া গেছে:")
    for ensg in top_ensg[:5]:  # প্রথম ৫টা দেখি
        if ensg in adata_full.var.index:
            symbol = adata_full.var.loc[ensg, 'gene_symbols']
            print(f"{ensg} → {symbol}")
```
